```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Average Income': [25000, 28000, 32000, 35000, 38000]  
}

df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
plt.plot(df['Average Income'], df['Year'], marker='o', color='g', label='Average Income')
plt.title('Average Income vs Year')
plt.xlabel('Average Income (USD)')
plt.ylabel('Year')
plt.grid(True)
plt.legend()

plt.show()
```
