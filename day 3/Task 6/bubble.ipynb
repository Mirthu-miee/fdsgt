```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
np.random.seed(42) 
n_countries = 8
years = [1995, 2000, 2005, 2010, 2015]

data = {
    'Year': [],
    'Country': [],
    'Life Expectancy': [],
    'Income': [],
    'Population': []
}

for year in years:
    for i in range(n_countries):
        country = f"Country {i+1}"
        life_expectancy = np.random.randint(60, 85)  
        income = np.random.randint(1000, 60000)      
        population = np.random.randint(1000000, 100000000)
        data['Year'].append(year)
        data['Country'].append(country)
        data['Life Expectancy'].append(life_expectancy)
        data['Income'].append(income)
        data['Population'].append(population)

df = pd.DataFrame(data)


def create_bubble_chart(year):
    df_year = df[df['Year'] == year]

    plt.figure(figsize=(12, 8)) 
    population_size = (df_year['Population'] / df_year['Population'].max()) * 500  

    plt.scatter(df_year['Income'], df_year['Life Expectancy'], s=population_size, alpha=0.7, c=df_year.index, cmap='viridis') 

    plt.xlabel('Income per Capita')
    plt.ylabel('Life Expectancy')
    plt.title(f'Life Expectancy vs. Income ({year})')
    plt.xscale('log')
    plt.colorbar(label='Country Index')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

create_bubble_chart(years[0])

try:
    import ipywidgets as widgets
    dropdown = widgets.Dropdown(options=years, value=years[0], description='Year:')
    widgets.interactive(create_bubble_chart, year=dropdown)
except ImportError:
    print("ipywidgets not found.  Install it for interactive dropdowns (pip install ipywidgets).")
    print("Showing static chart for", years[0]) 
```
