```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Happiness Score (WHR)': [5.5, 5.8, 6.0, 6.5, 6.8] 
}

df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
plt.plot(df['Happiness Score (WHR)'], df['Year'], marker='o', color='purple', label='Happiness Score (WHR)')
plt.title('Happiness Score (WHR) vs Year')
plt.xlabel('Happiness Score (WHR)')
plt.ylabel('Year')
plt.grid(True)
plt.legend()

plt.show()
```
